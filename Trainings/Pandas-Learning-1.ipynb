```python
df = pd.DataFrame()
df['col_1'] = [1,4,9,16]
df['col_2'] = [25,36,49,64]
df
```

```python
def sum_row(row):
    sum = row['col_1'] + row['col_2']
    return sum
df["Sum"] = df.apply(sum_row, axis=1)
df
```

```python
df.drop(['Sum'], axis=1, inplace=True)
df.apply(np.sum)
df
```

```python
def multiply_row(row):
    return row * 2
df.apply(multiply_row, axis=1)
```

```python
def divide_row(row):
    return row['col_2'] / row['col_1']
df['Divide'] = df.apply(divide_row, axis=1)
df
```

```python
df.drop(['Divide'], axis=1, inplace=True)
df
```
