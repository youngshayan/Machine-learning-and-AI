```python
import numpy as np
import matplotlib.pyplot as plt
def mandelbrot(h, w, maxit=20, r=2):
    """Returns an image of the Mandelbrot fractal of size (h,w)."""
    x = np.linspace(-2.5, 1.5, 4*h+1)
    y = np.linspace(-1.5, 1.5, 3*w+1)
    A, B = np.meshgrid(x, y)
    C = A + B*1j
    z = np.zeros_like(C)
    divtime = maxit + np.zeros(z.shape, dtype=int)
    for i in range(maxit):
        z = z**2 + C
        diverge = abs(z) > r                    # who is diverging
        div_now = diverge & (divtime == maxit)  # who is diverging now
        divtime[div_now] = i                    # note when
        z[diverge] = r                          # avoid diverging too much
    return divtime
plt.clf()
plt.imshow(mandelbrot(400, 400))
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('fivethirtyeight')

age = [22, 24, 26, 28, 30, 31, 35, 39, 40, 41, 46, 49]
x_index = np.arange(len(age))

salary = [20046, 17100, 20000, 24744, 30500, 37732, 41247, 45372, 48876, 53850, 57287, 63016]
salary_pydev = [16446, 16791, 18942, 21780, 25704, 29000, 34372, 37810, 43515, 46823, 49293, 69000]
salary_js = [37810, 43515, 46823, 49293, 53437, 56373, 62375, 66674, 68745, 68746, 74583, 20000]

width = 0.25 

plt.bar(x_index, salary, width=0.25, label='All Developer Salaries')

plt.bar(x_index - width, salary_js, width=0.25, label='Javascript Developer Salaries')

plt.bar(x_index + width, salary_pydev, width=0.25, color= 'b', label='Python Developer Salaries')

plt.xticks(ticks=x_index,labels=age)
plt.legend()
plt.title('Programmers median salary based on age')
plt.xlabel('Age')
plt.ylabel('Median Salary')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter

plt.style.use('fivethirtyeight')

data = pd.read_csv('program.csv')
responder_id = data['Responder_id']
language = data['LanguagesWorkedWith']

language_counter = Counter()
for x in language:
    language_counter.update(x.split(';'))
    
languages = []
resp_id = []
for item in language_counter.most_common(15):
    languages.append(item[0])
    resp_id.append(item[1])
    
# if we want most popular languages be at the top, we should use reverse func
languages.reverse()
resp_id.reverse()

plt.barh(languages, resp_id)
plt.title("Most popular Programming Languages")
plt.xlabel("number of people")

plt.show()
```

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

plt.pie(resp_id,labels=languages, startangle=-90, autopct='%1.1f%%')
plt.title("Most popular Programming Languages")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('data3.csv')

age = data['Age']
all_devs = data['All_Devs']
python = data['Python']
javascript = data['JavaScript']

plt.plot(age, all_devs, label='All Developer Salaries')
plt.plot(age, python, label='Python Developer Salaries', linestyle='--')
# plt.plot(age, javascript, label='JavaScript Developer Salaries')

# for above
plt.fill_between(age, python, all_devs, where = (python > all_devs), interpolate = True
                 ,alpha = 0.25, label = 'Above')

# for below
plt.fill_between(age, python, all_devs, where = (python < all_devs), interpolate = True
                 , color = 'red', alpha = 0.25, label = 'Below')


plt.legend()

plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('seaborn-v0_8-bright')

data = pd.read_csv('data6.csv')

data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)

date = data['Date']
price_close = data['Close']
price_open = data['Open']

plt.plot(date, price_close,  label='Close Price', marker='o', linestyle='solid', color='y')
plt.plot(date, price_open, label='Open Price', marker='o', linestyle='solid', color='red')
plt.gcf().autofmt_xdate()


plt.fill_between(date, price_open, price_close, where = (price_open < price_close), interpolate = True,
                color = 'green', alpha = 0.25, label = 'Above')

plt.fill_between(date , price_open, price_close, where = (price_open > price_close), interpolate = True
                 , color = 'red', alpha = 0.25, label = 'Below')

plt.legend()
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Bitcoin price')
plt.tight_layout()

plt.show()
```
